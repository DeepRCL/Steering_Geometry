```python
from pathlib import Path

import pandas as pd

INPUT_NAME = "dataset_positive_answer_mapped.csv"

# Resolve data dir for repo root, dataset_construction/, or value_bench/ as cwd
_cwd = Path.cwd()
_candidates = [
    _cwd / "dataset_construction" / "data" / INPUT_NAME,
    _cwd / "data" / INPUT_NAME,
    _cwd.parent / "data" / INPUT_NAME,
]
INPUT_CSV = next((p for p in _candidates if p.exists()), None)
if INPUT_CSV is None:
    raise FileNotFoundError(f"Could not locate {INPUT_NAME} from cwd={_cwd}")

DATA_DIR = INPUT_CSV.parent
print(f"Input: {INPUT_CSV.resolve()}")

df = pd.read_csv(INPUT_CSV)
print(f"Loaded {len(df)} rows, columns: {list(df.columns)}")
df.head()
```

```python
# Drop rows with missing or NA mapped_value, then full-row duplicates
before = len(df)
df_ok = df[df["mapped_value"].notna()].copy()
df_ok = df_ok[df_ok["mapped_value"].astype(str).str.strip().str.upper() != "NA"]
after_na = len(df_ok)

df_clean = df_ok.drop_duplicates()
after_dedup = len(df_clean)

print(f"After removing NA mapped_value: {after_na} rows (dropped {before - after_na})")
print(f"After drop_duplicates(): {after_dedup} rows (dropped {after_na - after_dedup})")

# Count per canonical mapped_value
counts = df_clean["mapped_value"].value_counts().sort_values(ascending=False)
counts
```

```python
OUTPUT_CSV = DATA_DIR / "dataset_positive_answer_mapped_clean.csv"
df_clean.to_csv(OUTPUT_CSV, index=False)
print(f"Saved {len(df_clean)} rows → {OUTPUT_CSV.resolve()}")
```
