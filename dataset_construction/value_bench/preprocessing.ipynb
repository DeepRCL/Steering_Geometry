```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('./data/value_bench.csv')
```

```python
values = df['value'].unique()
print(values)
```

```python
print(len(values))
counts = df['value'].value_counts()
print(counts)
```

```python
values = ['Power', 'Hedonism', 'Stimulation', 'Achievement', 'Self-Direction', 'Security', 'Tradition', 'Conformity', 'Humility', 'Benevolence', 'Universalism']
print(df['value'].value_counts().reindex(values, fill_value=0))
```

```python
df_filtered = df[df['agreement'] == 1]

df_dataset_positive = pd.DataFrame({
    'question': df_filtered['question'],
    'value': df_filtered['value'],
    'positive_answer': df_filtered['item'],
    'negative_answer': ''
    })

df_dataset_positive.head(10)
```

```python
df_filtered = df[df['agreement'] == -1]

df_dataset_negative = pd.DataFrame({
    'question': df_filtered['question'],
    'value': df_filtered['value'],
    'positive_answer': '',
    'negative_answer': df_filtered['item']
    })

df_dataset_negative.head(10)
```

```python
df_dataset_positive.to_csv('./data/dataset_positive_only.csv', index=False)
df_dataset_negative.to_csv('./data/dataset_negative_only.csv', index=False)
```
